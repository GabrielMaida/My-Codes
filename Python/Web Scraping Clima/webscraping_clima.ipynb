**Extração das temperaturas**

```python
temperatura = html.find('span', class_='dato-temperatura changeUnitT')
print(temperatura.text)

temp_max = html.find('span', class_='max changeUnitT')
temp_min = html.find('span', class_='min changeUnitT')
print(temp_max.text)
print(temp_min.text)
```

## **Código Completo**

```python
#!pip install requests
#!pip install xlsxwriter
#!pip install beautifulsoup4

import requests
#import xlsxwriter
from bs4 import BeautifulSoup

URL = 'https://www.tempo.com/ingleses-do-rio-vermelho.htm'
request = requests.get(URL)
html = BeautifulSoup(request.text, 'html.parser')

print("News of Python page")
print("Site:", URL)
title = html.title.text
formatted_title = title.split('.')[0].strip()
print("Title Page: ", formatted_title)

#print(html,'\n')
#print(request)

temperatura = html.find('span', class_='dato-temperatura changeUnitT')
print(temperatura.text)

temp_max = html.find('span', class_='max changeUnitT')
temp_min = html.find('span', class_='min changeUnitT')
print(temp_max.text)
print(temp_min.text)

data = html.find('span', class_='subtitle-m')
print(data.text)
```
