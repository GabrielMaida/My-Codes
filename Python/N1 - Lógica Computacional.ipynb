## Questão 3 - Tipos de Variáveis e Condições

```python
data = 3 + 4j
if type(data) == complex:
  result = data.real
else:
  result = "Not complex"
print(result)
```

**Resposta:** a) 3.0

**Justificativa:** Como a variável *data* é do tipo *complex*(complexo), o código executa o que está dentro do primeiro bloco do *if*, que no caso é o comando `result = data.real`, o qual pega a parte real do número complexo (3.0 no caso) e armazena na variável *result*. Assim, no final, o `print(result)` retorna ao usuário o valor 3.0 de tipo *float*.

## Questão 4 - Condições com Strings

```python
text = "PYTHON"
if text.lower() == "python":
  output = text[1:4]
else:
  output = "Mismatch"
print(output)
```

**Resposta:** e) nenhuma das respostas

**Justificativa:** Neste código a condição do *if* é `text.lower() == "python"`, que é o mesmo que verificar se a variável *text*, convertida em letras minúsculas, é igual a *string* "python". Neste caso ela é, então o que o bloco *if* faz é salvar na variável *output* o segundo, terceiro e quarto caracter da *string text* através do comando `text[1:4]`, pois o primeiro caracter sempre é o 0. Finalmente, após a condicional é printado na tela a variável *output*, que é igual a *string* "YTH", isso, porque o valor salvo em *output* é o valor inicial da variável *text*, ou seja, PYTHON em letras maiúsculas.

## Questão 6 - Uso de Operadores de Comparação

```python
x = 10
y = 20
if x * 2 >= y:
  output = "Verdadeiro"
else:
  output = "Falso"
print(output)
```

**Resposta:** a) Verdadeiro

**Justificativa:** A saída é "Verdadeiro", porque x * 2 é igual a 20 e condição do *if* de `x * 2 >= y` é verdadeira. Assim, a *string* "Verdadeiro" é salva na variável e printada ao usuário.

## Questão 8 - Condições Encadeadas

**Resposta:** b) Neutro

## Questão 10 - Operações Matemáticas e Condições

```python
value = 25
if value**0.5 == 5:
    output = "Quadrado Perfeito"
else:
    output = "Não é Quadrado Perfeito"
print(output)
```

**Resposta:** a) Quadrado Perfeito

**Justificativa:** O que o trecho de código apresentado faz é armazenar o número 25 na variável *value*, após isso, o *if* verifica se a condição `value ** 0.5` é igual a 5, ou seja, se 25 elevado a 0.5 é igual a 5. Como essa condição é verdadeira, a variável *output* recebe a *string* "Quadrado Perfeito", que é printada na tela com o `print(output)`.
